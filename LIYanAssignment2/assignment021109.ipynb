```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python
class Film:
    def __init__(self,Name,Number,Normal):
        self.Name = Name
        self.Number = Number
        self.Normal = Normal
```

```python
with open("PokeTypeMatchupData.csv")as csvfile:
    reader = csv.reader(csvfile)
    print(next(reader))
```

```python
with open("PokeTypeMatchupData.csv")as csvfile:
    reader = csv.reader(csvfile)
    rows = 0
    for r in reader:
        rows +=1
    print(rows)    
```

```python
Films = []
with open("PokeTypeMatchupData.csv")as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        Name = row['Name']
        Number = row['Number']
        Normal = row['Normal']
        Films.append(Film(Name,Number,Normal))
```

```python
print(Films[0].Name)
```

```python
for c in Films[0].Name:
    print(c)
```

```python
FilmGraph = nx.Graph()
```

```python
filmSubset = Films[0:20]
```

```python
for f in filmSubset:
    FilmGraph.add_node(f)
```

```python
for f1 in filmSubset:
    for f2 in filmSubset:
        if not f1 == f2:
            if f1.Number == f2.Number:
                FilmGraph.add_edge(f1,f2)
            for c in f2.Number:
                FilmGraph.add_edge(f1,f2)
```

```python
options = {'node_color':'black','node_size':10,'width':1}
plt.figure(1,figsize = (30,30))
nx.draw(FilmGraph,**options)
```
